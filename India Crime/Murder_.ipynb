# No. of Murders in India from 2001 to 2010

```python
lst = []
for i in df['Year'].unique():
    yr = df[(df['Year']==i)] 
    lst.append(sum(yr['Victims_Total']))
```

```python
plt.figure(figsize = (10,6))
plt.plot( df['Year'].unique(),lst)
plt.xlabel('Year')
plt.ylabel('Murder')
plt.show()
```
