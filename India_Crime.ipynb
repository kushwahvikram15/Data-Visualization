```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('20_Victims_of_rape.csv')
df.head()
```

### Rape_Cases_Reported and Victims_of_Rape_Total are the same so we remove the redundant column

```python
df = df.drop('Rape_Cases_Reported',axis = 1)
```

# Number of rapes with respect to each state in 2001

```python
x = df[(df['Year']==2001) & (df['Subgroup']=='Total Rape Victims')]
plt.figure(figsize = (14,9))
plt.bar(x['Area_Name'],x['Victims_of_Rape_Total'])
plt.title('No. of Rapes with respect to each state')

plt.xticks(rotation = 90)
plt.show()
```

 ### Madhya pradesh leads in the number of rape cases followed by uttar pradesh and maharashtra

```python
max_case = df[df['Victims_of_Rape_Total']==max(df['Victims_of_Rape_Total'])]
print('The Most Unsafe state in India for women',max_case['Area_Name'].values[0])
```

```python
min_case = df[df['Victims_of_Rape_Total']==min(df['Victims_of_Rape_Total'])]
print('The Most Safest state in India for women is',df['Area_Name'].values[0])
```

# Number of rapes in India from 2001 to 2010

```python
lst = []
for i in df['Year'].unique():
    x = df[(df['Year']==i)& (df['Subgroup']=='Total Rape Victims')]
    s = sum(x['Victims_of_Rape_Total'])
    lst.append(s)
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['Year'].unique(),lst,color = 'red',marker = '*')
plt.title('No. of rapes with respect to year')
plt.xlabel('No.of rapes')
plt.ylabel('Year')
plt.plot()
```
