```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data_set = pd.read_csv('Netflix Shows.csv', encoding = "ISO-8859-1")
```

```python
data_set.tail()
```

### There is lot of data duplicates in title so we need to findout those data in title and remove them.

```python
data_set.info()
```

```python
data_set[data_set.duplicated('title',keep = False)]
```

```python
data_set = data_set.drop_duplicates()
data_set.info()
```

### Now there is no such data are duplicate in title.

```python
data_set['release year'].unique()
```

so there is clear to all movies released from the year 1987-2017

```python
plt.rcParams['axes.labelsize']= 25
plt.rcParams['axes.edgecolor'] = 'red'
plt.rcParams['axes.facecolor'] = '#ceebd6'
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False
plt.rcParams['figure.dpi'] = 200
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.rcParams['axes.titlesize'] = 30
plt.rcParams['axes.titlepad'] = 20
plt.rcParams['grid.linestyle'] = 'dashdot'
```

```python
yr = data_set.groupby('release year').size()
plt.figure(figsize = (10,6))
plt.plot(yr.index, yr.values, color = 'red')
plt.xlabel('Years')
plt.ylabel('No. of Movies')
plt.title('Number of shows released with respect to each year')
plt.show()
```

### The number of shows almost increases every year apart from a few fluctuations.
### An exponential increase is seen from the year 2015 to 2016 (We don't have enough data for 2017)

```python
print('Number of shows for families to watch together:',data_set[(data_set['ratingLevel']=='Suitable for all ages.') | (data_set['ratingLevel']=='General Audiences. Suitable for all ages.') ]['title'].count())
```

```python
print('Number of shows for Children:',data_set[(data_set['ratingLevel']=='Suitable for children ages 7 and older') | (data_set['ratingLevel']=='Suitable for children ages 7 and older.  Content contains fantasy violence.')]['title'].count())
```

```python
print('Number of unrated shows:',data_set[(data_set['ratingLevel']=='NaN') | (data_set['ratingLevel']=='This movie has not been rated.') ]['title'].count())
```

# The Top Rated Show on Netflix

```python
top_rated = data_set[data_set['user rating score']>90]['title']
top = top_rated.values

for i in range(0, len(top)):
    print(top[i])
```

# Latest Top Rated show on Netflix

```python
latest_best = data_set[(data_set['user rating score']>90) & (data_set['release year']>2016)]['title']
best = latest_best.values
for i in range(0, len(best)):
    print(best[i])
```

```python
rat = data_set['rating'].value_counts()
sns.barplot(rat.index, rat.values, alpha=0.8)
plt.xticks(rotation='vertical')
plt.xlabel('Type of rating', fontsize=14)
plt.ylabel('Number os shows', fontsize=14)
plt.title("Rating of Netflix programs", fontsize=16)
plt.show()
```
